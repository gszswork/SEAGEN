```python
import os
```

```python
category = os.listdir("./all-rnr-annotated-threads")
```

```python
label_dict = {} # 0 non-rumour, 1 rumour
for cate in category:
    rumour_ids = os.listdir("./all-rnr-annotated-threads/"+cate+"/"+"rumours")
    non_rumour_ids = os.listdir("./all-rnr-annotated-threads/"+cate+"/"+"non-rumours")
    
    rumour_ids = [i for i in rumour_ids if i[0]!='.']
    non_rumour_ids = [i for i in non_rumour_ids if i[0]!='.']
    
    for i in rumour_ids:
        label_dict[i] = 1
    for i in non_rumour_ids:
        label_dict[i] = 0
    
```
